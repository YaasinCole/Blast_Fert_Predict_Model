```python
##Implementing beta negative likelihood (beta regression)
import tensorflow as tf
from tensorflow.keras import backend as K

def beta_nll_loss(y_true, y_pred):
    """
    Beta Negative Log-Likealihood Loss for regression of proportions (0 < y < 1).
    Assumes the model outputs two values per target:
      - mean (mu): predicted mean of the beta distribution (01)
      - precision (phi): controls the shape/variance (>0)
    """
    # Split model output into mean (mu) and precision (phi)
    mu = tf.clip_by_value(y_pred[:, 0], 1e-6, 1 - 1e-6)     # avoid log(0)
    phi = tf.nn.softplus(y_pred[:, 1]) + 1e-6               # ensure phi > 0

    # Compute alpha and beta
    alpha = mu * phi
    beta = (1 - mu) * phi

    # Beta log-likelihood
    nll = -(
        tf.math.lgamma(alpha + beta)
        - tf.math.lgamma(alpha)
        - tf.math.lgamma(beta)
        + (alpha - 1) * tf.math.log(y_true)
        + (beta - 1) * tf.math.log(1 - y_true)
    )

    return tf.reduce_mean(nll)
```

```python
from tensorflow.keras.callbacks import EarlyStopping, ModelCheckpoint, ReduceLROnPlateau
from tensorflow.keras import regularizers
import math
from sklearn.metrics import mean_absolute_error, mean_squared_error

def evaluate_percent(y_true_scaled , p_pred_mu, name="model"):
# the y_true_scaled is in the 0-1 range we used in training as our target 
# y_pred_mu : the model predictions of(0-1 range)

    y_true_pct = y_true_scaled *100.0
    y_pred_pct = p_pred_mu  *100.0
    mae = mean_absolute_error(y_true_pct, y_pred_pct)
    rmse = math.sqrt(mean_squared_error(y_true_pct, y_pred_pct))
    print(f"{name}  | MAE (pct): {mae:.3f}  | RMSE (pct): {rmse:.3f}")
    return mae, rmse
```

```python
#BUILD NEURAL NETWORK 
import tensorflow as tf 
from tensorflow import keras 
from tensorflow.keras import layers, Sequential, optimizers
from tensorflow.keras.layers import BatchNormalization, LayerNormalization
from tensorflow.keras.metrics import AUC
from sklearn.metrics import classification_report, confusion_matrix , roc_curve
from sklearn.utils.class_weight import compute_class_weight

#Clear any previous sessions 
tf.keras.backend.clear_session()

def mae_mu(y_true, y_pred):
    mu = y_pred[:, 0]  # only compare predicted mu
    return tf.reduce_mean(tf.abs(y_true - mu))

HP = {
    "lr": 5e-4,                 # try 1e-3, 3e-4, 1e-4
    "weight_decay": 1e-5, 
    "l2_reg": 1e-5,            # try 1e-6 .. 1e-4
    "units1": 128,             # try 64, 128, 256
    "units2": 64,              # try 32, 64
    "dropout": 0.05,           # try 0.05, 0.1, 0.2
    "batch_size": 16,          # try 16 or 32
    "patience_es": 12,         # EarlyStopping patience
    "patience_rlr": 6,         # ReduceLROnPlateau patience
}

def create_beta_model(input_dim, model_name,hp=HP):
        inputs = layers.Input(shape=(input_dim,), name=f'{model_name}_input')
        
        x = layers.Dense(hp["units1"], use_bias=False, kernel_regularizer=tf.keras.regularizers.l2(hp["l2_reg"]),
                         name=f'{model_name}_hidden1')(inputs)
        x = BatchNormalization()(x)
        x = layers.LeakyReLU(alpha=0.1)(x)
        x = layers.Dropout(hp["dropout"])(x)

        # x = layers.Concatenate()([x, inputs])

        x = layers.Dense(hp["units2"], use_bias=False,kernel_regularizer=tf.keras.regularizers.l2(hp["l2_reg"]),
                         name=f'{model_name}_hidden2')(x)
        x = BatchNormalization()(x)
        x = layers.LeakyReLU(alpha=0.1)(x)
        x = layers.Dropout(hp["dropout"])(x)
        
        # mu_raw = layers.Dense(1, activation=None, name=f'{model_name}_mu_raw')(x)
        mu = layers.Dense(1, activation='sigmoid', name=f'{model_name}_mu')(x)
        phi = layers.Dense(1, activation='softplus', name=f'{model_name}_phi')(x)

        outputs = layers.Concatenate(name=f'{model_name}_output')([mu, phi])

        model = keras.Model(inputs, outputs, name=f'{model_name}_beta_model')
        optimizer = tf.keras.optimizers.AdamW(learning_rate=hp["lr"], weight_decay=hp["weight_decay"])
    
    
        model.compile(
        optimizer=optimizer,
        loss=beta_nll_loss,
        metrics=[mae_mu]
        )
    
        return model 

# Get number of classes
# num_fert_classes = len(np.unique(y_fert_train))
# num_blast_classes = len(np.unique(y_blast_train))

# print(f"Number of fertilization classes: {num_fert_classes}")
# print(f"Number of blastulation classes: {num_blast_classes}")

#Setting the training data to the input for our MLP model,
input_dim = x_train_scaled.shape[1]

fert_model = create_beta_model(input_dim, "fert")
blast_model = create_beta_model(input_dim, "blast")

#Outputting what the model looks like and what is going to be used in training
print("Fertilization Model Summary:")
fert_model.summary()
print("\nBlastulation Model Summary:")
blast_model.summary()
```

```python
from tensorflow.keras.callbacks import EarlyStopping, ModelCheckpoint, ReduceLROnPlateau
from sklearn.metrics import f1_score
import numpy as np
import tensorflow as tf

#implementing a custom F1 Score method to use with the model
# class F1ScoreCallback(tf.keras.callbacks.Callback):
#     def __init__(self ,x_val , y_val, threshold=0.5):
#         super().__init__()
#         self.x_val = x_val
#         self.y_val = y_val
#         self.threshold = threshold 
#         self.best_f1 = 0

#     def on_epoch_end(self, epoch, logs=None):
#         ##predicting on the validation set 
#         y_pred = self.model.predict(self.x_val, verbose=0)

#         ##Threshold predictions to binary 
#         y_pred_class =(y_pred >= self.threshold).astype(int)
#         y_true_class= (self.y_val >= self.threshold).astype(int)

#         ##Computing the f1Score 
#         f1 = f1_score(y_true_class , y_pred_class)

#         logs['f1_score'] =f1
#         print(f" - val_f1 : {f1:.4f}")

#         ##Saving the best model 
#         if f1> self.best_f1:
#             self.best_f1 = f1 
#             self.model.save(f"best_model_f1_{self.best_f1:.4f}.keras")

# f1_callback_fert = F1ScoreCallback(x_val_scaled, y_fert_val)
# f1_callback_blast = F1ScoreCallback(x_val_scaled, y_blast_val)
##Implementing EarlyStopping and ModelCheckpoint
# For us to save the best model and stop the model when it stops improving
fert_callbacks = [
    EarlyStopping(monitor='val_loss', patience=HP["patience_es"], restore_best_weights=True, verbose=1),
    ModelCheckpoint('best_fert_model.keras', save_best_only=True, monitor='val_loss'),
    ReduceLROnPlateau(monitor='val_loss', factor=0.7, patience=HP["patience_rlr"], min_lr=1e-5)
    # f1_callback_fert
]

blast_callbacks = [
    EarlyStopping(monitor='val_loss', patience=HP["patience_es"], restore_best_weights=True, verbose=1),
    ModelCheckpoint('best_blast_model.keras', save_best_only=True, monitor='val_loss'),
    ReduceLROnPlateau(monitor='val_loss', factor=0.7, patience=HP["patience_rlr"], min_lr=1e-5)
    # f1_callback_blast
]
```

```python
##Training the models 

print("Training Fertilization Model...")
fert_history = fert_model.fit(
    x_train_scaled, y_fert_train,
    validation_data=(x_val_scaled, y_fert_val),
    epochs=200,
    batch_size=HP["batch_size"],
    callbacks=fert_callbacks,
    verbose=1
)

print("\nTraining Blastulation Model...")
blast_history = blast_model.fit(
    x_train_scaled, y_blast_train,
    validation_data=(x_val_scaled, y_blast_val),
    epochs=200,
    batch_size=HP["batch_size"],
    callbacks=blast_callbacks,
    verbose=1
)
```
